```python
# Add column that converts venue to categorical column where "0" indicates "Away" and "1" indicates "Home"
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
# Add column that converts opponent to categorical column where each opponent is represented by a number
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
# 0: Draw, 1: Lose, 2: Win
# matches["target"] = matches["result"].astype("category").cat.codes

matches["target"] = (matches["result"] == "W").astype("int")
matches.target
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
X_train = matches[matches["date"] < "2023-01-01"]
X_test = matches[matches["date"] > "2023-01-01"]
# matches.date.dt.year.value_counts()
```

```python
X_train.shape, X_test.shape
```

```python
matches.columns
```

```python
features = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rfc.fit(X_train[features], X_train["target"])
```

```python
preds = rfc.predict(X_test[features])
```

```python
preds
```

```python
pd.concat([X_test["target"].reset_index(), pd.DataFrame(preds)], axis=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(X_test["target"], preds)
```
